```python
# afrique 
afr = pd.read_csv('bilan_nations_afrique_hommes.csv')
afr.head(20)
```

```python
# Modifier les en-têtes en passant les dates de YY à YYYY

# on selectionne la 1e ligne avec les entetes
years=afr.columns

# on va modifier pour les valeurs numeriques les anne yy en yyyy. et on distingue 19yy e 20yy
new_years=[]
for year in years:
    if year.isdigit(): 
        if 0 <= int(year) <= 50:
            new_year = '20' + year 
        else:
            new_year = '19' + year
    else:
        new_year=year
    new_years.append(new_year)

print(new_years)
afr.columns = new_years
afr.head(20)
afr.shape
```

```python
#A revoir - essai modifier les float en int. mais a revoir ou autre solution

select=afr.iloc[0:,1:]
print(select)


def convert_to_int(value):
    try:
        return int(value)  # Convertir en float d'abord pour gérer les nombres décimaux sous forme de chaîne
    except ValueError:
        return value  # Retourner la valeur inchangée si la conversion échoue

# Appliquer la conversion à chaque élément des colonnes sélectionnées
afr.iloc[1:, 1:] = select.applymap(convert_to_int)

#for value in select:
#    afr[col]=afr[col].astype(int)

afr.head(20)
```
