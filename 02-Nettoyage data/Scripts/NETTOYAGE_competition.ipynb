```python
import pandas as pd 
import csv
import numpy as np
```

Afrique 

```python
# afrique 
afrique = pd.read_csv('bilan_nations_afrique_hommes.csv')
afrique.head(20)
```

```python
# resoudre le pb des donnes qui ont ete converties en float lors de l ouverture du fichier. du coup on cherche a les reconvertir en int. 
# il faut d abord remplacer les valeurs non numeriques comme Nan / disqualifiée / ou '_' et on les remplace par 0

afrique.replace("Forfait",np.nan,inplace=True)
afrique.replace("Disqualifié",np.nan,inplace=True)
afrique.replace("-",np.nan,inplace=True)
afrique.fillna(0,inplace=True)
# print(afrique.to_string())

for i in afrique.columns[1:]:
    afrique[i] = afrique[i].astype(int)
print(afrique.to_string())

```

```python
# Modifier les en-têtes en passant les dates de YY à YYYY

# on selectionne la 1e ligne avec les entetes
years=afrique.columns

# on va modifier pour les valeurs numeriques les anne yy en yyyy. et on distingue 19yy e 20yy
new_years=[]
for year in years:
    if year.isdigit(): 
        if 0 <= int(year) <= 50:
            new_year = '20' + year 
        else:
            new_year = '19' + year
    else:
        new_year=year
    new_years.append(new_year)

print(new_years)
afrique.columns = new_years
afrique.head(20)
```

```python
# sauvegarde dans un autre nouveau fichier csv
afrique.to_csv('bilan_nations_afrique_hommes_clean.csv')
```

```python
amsud = pd.read_csv('bilan_nations_AmSud_hommes.csv')
amsud.head(20)
```

```python
# Pour homogeneiser entre toutes les compétitions. Renommer la colonne 'Équipe nationale' en 'Nation'
amsud.rename(columns={"Équipe nationale": "Nation"}, inplace=True)
amsud.head(20)
```

```python
#remplacement 99 par Nan
amsud.replace('99', np.nan, inplace=True)
amsud.head(30)
```

```python
#remplacement 4e par 4 , 5e par 5 , 6e, Nan
amsud.replace({'01':1, '02':2,'03':3, '4e': 4, '5e': 5, '6e': 6, '7e':7}, inplace=True)
amsud.head(20)
```

```python
#remplacement 88 par Nan  - est ce qu on doit remplacer 
amsud.replace('88', np.nan, inplace=True)
amsud.head(30)
```

```python
#modifie tous les nan en 0
amsud.fillna(0,inplace=True)
amsud.head(10)
```

```python
# sauvegarde dans un autre nouveau fichier csv
amsud.to_csv('bilan_nations_AmSud_hommes_clean.csv')
```
